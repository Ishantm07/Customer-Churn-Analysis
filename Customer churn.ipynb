```python
# converted blank values with zero and datatype to float
df['TotalCharges']=df['TotalCharges'].replace(" ",0)
df['TotalCharges']=df['TotalCharges'].astype('float')
```

```python
df.info()
```

```python
df['customerID'].duplicated().sum()
```

```python
# converted 0 and 1 values in Senior citizen to yes and no
def conv(value):
    if value ==1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head(5)
```

```python
ax = sns.countplot(x='Churn', data=df)

# Add labels to each bar
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Count of customer by churn")
plt.show()
```

```python
plt.figure(figsize=(3,4))
gd = df.groupby('Churn').agg({'Churn': "count"})

plt.pie(gd['Churn'],labels=gd.index, autopct = "%1.1f%%")
plt.show()
```

### From the above chart we found out that 26.5% of the customers have churned

```python
plt.figure(figsize=(5,3))
ax = sns.countplot(x = "gender",hue = "Churn",data = df)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Gender wise Churn")
plt.show()
```

```python
plt.figure(figsize=(3,4))
gd = df.groupby('SeniorCitizen').agg({'SeniorCitizen': "count"})

plt.pie(gd['SeniorCitizen'],labels=gd.index, autopct = "%1.1f%%")
plt.show()
```

```python

count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

count_data_percentage = count_data.div(count_data.sum(axis=1), axis=0) * 100

plt.figure(figsize=(4, 4))
count_data_percentage.plot(kind='bar', stacked=True)

for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().text(p.get_x() + p.get_width() / 2., p.get_y() + height / 2., f'{height:.1f}%', ha='center')

plt.title("Senior Citizen wise Churn (Percentage of Total)")
plt.ylabel("Percentage")
plt.xlabel("Senior Citizen")
plt.show()
```

#### As per the graph 58% of the senior citizens are already been churned

```python
plt.figure(figsize=(9, 4))
sns.histplot(x = 'tenure',hue = 'Churn', data = df, bins =72)
plt.show()
```

#### People wo ave used our service for a long time have stayed but people who have used our service only for one month ahve churned

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = 'Contract',hue = 'Churn', data = df)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Count of Customers by Contract")    
plt.show()
```

#### Customers having month-to-month have mostly churned 

```python
df.columns
```

```python



columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))

axes = axes.flatten()

for i, col in enumerate(columns):
    sns.countplot(x=col, hue='Churn', data=df, ax=axes[i])
    axes[i].set_title(f'{col} vs Churn')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

#### Customers with MultipleLines, fiber optic, streaming services, and online backup have churned a lot.

```python
plt.figure(figsize = (9,4))
ax = sns.countplot(x = 'PaymentMethod',hue = 'Churn', data = df)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Count of Customers by PaymentMethod") 
plt.rotation = 90
plt.show()
```
